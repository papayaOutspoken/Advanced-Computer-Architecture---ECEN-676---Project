```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.distributions import Categorical
from g_share import GShare

LOG2_GHIST_LEN = 12
GHIST_LEN = 2**LOG2_GHIST_LEN
class HashModel(nn.Module):
    def __init__(self):
        super(HashModel, self).__init__()
        # lin_layers = [nn.Linear(2**(i)+(64 if i == LOG2_GHIST_LEN else 0),2**(i-1)) for i in reversed(range(6,LOG2_GHIST_LEN+1))]
        # lin_layers = [nn.Linear(2**12+64, 2**12), nn.Linear(2**12+64, 2**12)]
        lin_layers = [nn.Linear(16385, 2**12), nn.Linear(2**12,2**12)]
        layers= []
        for i in range(len(lin_layers)):
            layers.append(lin_layers[i])
            if i < len(lin_layers)-1:
                layers.append(nn.ReLU())
        self.layers = nn.Sequential(*layers)

    def forward(self, input: torch.Tensor) -> torch.Tensor:
        logits = self.layers(input)
        #return probability distribution
        return logits
    
```

```python
model = HashModel()
# https://www.reddit.com/r/MachineLearning/comments/qq75zu/d_how_do_you_choose_an_optimizer_and_why_are/
optim = lambda *ar, **kw: torch.optim.AdamW(*ar, **kw, amsgrad=True) 
# optim = lambda *ar, **kw: torch.optim.SGD(*ar, **kw, nesterov=True, momentum=MAGIC_NUM) 
optim = optim(model.parameters())
(optim, model)
```

```python
# gshare = GShare(hashfn=lambda addr, hist: hash((addr, hist)))
gshare = GShare()
branch_address_set = set()
correct_predict = []
correct_predict_cnt = 0
for (pc, taken) in branch_trace:
  # taken = not taken
  branch_address_set.add(pc)
  predict = gshare.predict_branch(pc)
  if ((predict >= 0 if taken else predict < 0)):
    correct_predict.append(1)
    correct_predict_cnt += 1
  else:
    correct_predict.append(0)
  gshare.update_predictor(pc, 1 if not taken else 0)
(
  len(branch_address_set),
  len(branch_trace),
  len(list(filter(lambda x: x[1], branch_trace))),
  correct_predict_cnt,
  list(zip(branch_trace, correct_predict)),
)
```

```python
def int_to_bitvector(n, num_bits):
    powers = 2 ** torch.arange(num_bits - 1, -1, -1)
    return (n & powers) > 0
def training_loop(
        model: nn.Module,
        predictor: GShare,
        optimizer: torch.optim.Optimizer,
        pc_stream: list,
        canonical_list: list):
    #train once
    model.train()
    total_loss = 0.0
    for i in range(min(5, len(pc_stream))):
        pc_val = torch.tensor([pc_stream[i]], dtype=torch.float32)
        history_val = int_to_bitvector(predictor.hist_vector, 16384)
        #selected the index and get the probability
        probabilities = F.softmax(model(torch.cat([history_val, pc_val])), dim=-1)
        distribution = Categorical(probabilities)
        selected_index = distribution.sample()
        log_prob = distribution.log_prob(selected_index)
        #poll the branch predictor
        predictor_input = selected_index.item()
        predictor_output = predictor.predict_branch(predictor_input)
        correct_direction = not canonical_list[i]
        predictor.update_predictor(predictor_input, correct_direction)
        #calculate loss
        print(f"correct_direction={correct_direction}, predicted_output={predictor_output}")
        target = (float(correct_direction) * 2.0) - 1.0
        mse_loss_val = (predictor_output - target) ** 2       
        grad_loss = torch.tensor(mse_loss_val, requires_grad=False) * -log_prob
        #back propagation
        optimizer.zero_grad()
        grad_loss.backward()
        optimizer.step()
        print(f"loss={mse_loss_val}")
        total_loss += mse_loss_val
        
    average_mse = total_loss/len(pc_stream)
    return average_mse
```

```python
pc_stream = list(map(lambda x: x[0], branch_trace))
canonical_list = list(map(lambda x: x[1], branch_trace))
```

```python
training_loop(model, gshare, optim, pc_stream, canonical_list)
```
